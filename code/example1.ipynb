## Определим код диаграммы

```python
CODE = """
digraph structs {
    node [shape=record];
    struct1 [label="<f0> left|<f1> mid&#92; dle|<f2> right"];
    struct2 [label="<f0> one|<f1> two"];
    struct3 [label="hello&#92;nworld |{ b |{c|<here> d|e}| f}| g | h"];
    struct1:f1 -> struct2:f0;
    struct1:f2 -> struct3:here;
}
"""
```

## Самый простой вариант - сразу отобразим результат:

```python
Diagram.display_code("graphviz", CODE)
```

При изменении кода нужно повторно выполнить предыдущую строку

## Более сложный вариант - для работы с Python вариантом, например:

```python
diagram = Diagram()
d = diagram.init_draw("mermaid.state")
```

Определим диаграмму:

```python
d.l(None, "A")
s_fork = d.fork()
d.l("A",s_fork)
d.l(s_fork, "B")
d.l(s_fork, "C")
s_join = d.join()
d.l("B",s_join)
d.l("C",s_join)
d.l(s_join, "D")
d.l("D", None)
```

Просто вызовем объект диаграммы что бы отобразить его:

```python
d
```

Альтернативный вариант:

```python
diagram.display(d)
```
